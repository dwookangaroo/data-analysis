```python
# scatter 그래프 그리기
# 인구수 별 cctv 개수
plt.figure(figsize=(6,6))
plt.scatter(data_result['인구수'], data_result['소계'], s=50) # x축 인구수 y축 소계, 마커사이즈 50
plt.xlabel('인구수')
plt.ylabel('CCTV')
plt.grid()
plt.show()
```

```python
fp1 = np.polyfit(data_result['인구수'], data_result['소계'], 1) # x값 인구수, y값 소계, 1차함수(y = ax + b)
```

```python
fp1
```

```python
type(fp1)
```

```python
f1 = np.poly1d(fp1) # 1차함수로 변경
fx = np.linspace(100000, 700000, 100)
```

```python
plt.figure(figsize=(5,5))
plt.scatter(data_result['인구수'], data_result['소계'], s=50)
plt.plot(fx, f1(fx), ls='dashed', lw=3, color='g') # x 는 fx, y는 f1(fx)
plt.xlabel('인구수')
plt.ylabel('CCTV')
plt.grid()
plt.show()
```

```python
# 오차 컬럼 만들기
# 오차 컬럼 = 절대값(소계 - 1차함수 x값에 인구수넣은 y값(소계))
data_result['오차'] = np.abs(data_result['소계'] - f1(data_result['인구수'])) #abs() 절대값

#df_sort는 오차큰순서대로 표
df_sort = data_result.sort_values(by='오차', ascending=False)
df_sort.head()
```

```python
# visualizing
plt.figure(figsize=(14, 10))
plt.scatter(data_result['인구수'], data_result['소계'], c=data_result['오차'], s=50) # c는 컬러값
plt.plot(fx, f1(fx), ls='dashed', lw=3, color='g')

for n in range(10):
    plt.text(df_sort['인구수'][n] * 1.02, df_sort['소계'][n] * 0.98, df_sort.index[n], fontsize=15)
                    # x=인구수중에 n번인덱스, y=소계중에 n번인덱스, s=n번인덱스이름
plt.xlabel('인구수')
plt.ylabel('인구당비율')
plt.colorbar()
plt.grid()
plt.show()
```
