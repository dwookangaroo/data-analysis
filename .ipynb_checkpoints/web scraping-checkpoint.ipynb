```python
from urllib.parse import urljoin
rank = [] #순위
main_menu = [] #메뉴이름
cafe_name = [] #가게이름
url_add = [] #url

url_base = 'http://www.chicagomag.com'

list_soup = soup.find_all('div', 'sammy')

for item in list_soup:
    rank.append(item.find(class_ ='sammyRank').get_text())
    
    tmp_string = item.find(class_ ='sammyListing').get_text()
    main_menu.append(re.split(('\n|\r\n'), tmp_string)[0])
    cafe_name.append(re.split(('\n|\r\n'), tmp_string)[1])
    # a 안에 href 안에 주소
    # http://www.chicagomag.com 가 있는게 있고 없는게 있다
    # urljoin 은 있으면 냅두고 없으면 붙여줌
    url_add.append(urljoin(url_base, item.find('a')['href']))
    
```

```python
#pandas DataFrame 만들기
data = {'Rank':rank, 'Cafe':cafe_name, 'Menu':main_menu, 'URL':url_add}
df = pd.DataFrame(data)
df.head()
```

```python
from tqdm.notebook import tqdm
```

```python
# 어느정도 진행이되고있는지 보여준다
import time
item_total = tqdm(df.index) #반복문에 돌려줄것을 넣어준다
price = []
address = []

for i in item_total:
    url = df['URL'][i]
    req = Request(url, headers=header)
    html = urlopen(req)
    soup = BeautifulSoup(html, 'html.parser')
    
    line_data = soup.find('p', 'addy').get_text()
    price.append(line_data.split()[0][:-1])
    address.append(' '.join(line_data.split()[1:-2]))
    time.sleep(0.05)
```

```python

lat = []
lng = []

eter_val = tqdm(df.index)
for n in eter_val:
    if df['Address'][n] != 'Multiple':
        target_name = df['Address'][n] + ' Chicago'
        gmaps_output = gmaps.geocode(target_name)
        location_output = gmaps_output[0].get('geometry')
        lat.append(location_output['location']['lat'])
        lng.append(location_output['location']['lng'])
    else:
        lat.append(np.nan)
        lng.append(np.nan)
```
