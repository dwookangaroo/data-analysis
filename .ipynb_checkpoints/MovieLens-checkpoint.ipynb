```python
# 3개의 데이터 불러오기
import pandas as pd

pd.options.display.max_rows = 10

unames = ['user_id', 'gender', 'age', 'occupation', 'zip']
users = pd.read_table('data/movielens/users.dat', sep='::', header = None, names=unames)

rnames = ['user_id', 'movie_id', 'rating', 'timestamp']
ratings = pd.read_table('data/movielens/ratings.dat', sep='::', header = None, names=rnames)

mnames = ['movie_id', 'title', 'genres']
movies = pd.read_table('data/movielens/movies.dat', sep='::', header = None, names=mnames)
```

```python
users[:5]
```

```python
ratings[:5]
```

```python
movies[:5]
```

```python
# 데이터 합치기 merge
data = pd.merge(pd.merge(ratings, users), movies)
```

```python
data
```

```python
# 성별에 따른 각 영화의 평균 평점 구하기
# pivot table
#mean_ratings = data.pivot_table('rating', index = 'title', columns='gender', aggfunc = 'mean')
mean_ratings = pd.pivot_table(data, values = 'rating', index = 'title', columns='gender', aggfunc = 'mean')
```

```python
mean_ratings
```

```python
# 평점정보가 250건 이상있는 영화목록만 가져와보자
ratings_by_title = data.groupby('title').size() # title로 그룹핑하고 그 size
```

```python
ratings_by_title
```

```python
active_titles = ratings_by_title.index[ratings_by_title>=250]
```

```python
active_titles
```

```python
mean_ratings = mean_ratings.loc[active_titles]
```

```python
mean_ratings
```

```python
# 여성에게 높은 평점 받은순서
top_female_ratings = mean_ratings.sort_values(by='F', ascending=False)
```
